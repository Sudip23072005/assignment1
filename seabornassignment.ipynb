Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Here are five plots that can be created using the Seaborn library, along with their uses:

1. **Line Plot**:
   - **Usage**: `sns.lineplot()`
   - **Use**: Line plots are used to display data points connected by straight lines, commonly used to show trends over time or a continuous sequence. They are effective for visualizing the relationship between two continuous variables.

   ```python
   import seaborn as sns
   sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
   ```

2. **Bar Plot**:
   - **Usage**: `sns.barplot()`
   - **Use**: Bar plots represent data with rectangular bars with lengths proportional to the values they represent. They are used to compare different categories or groups, making it easy to see which category has the highest or lowest value.

   ```python
   sns.barplot(x="day", y="total_bill", data=tips)
   ```

3. **Scatter Plot**:
   - **Usage**: `sns.scatterplot()`
   - **Use**: Scatter plots display individual data points plotted in two-dimensional space, often used to observe relationships between two continuous variables. They help in identifying correlations, clusters, and outliers.

   ```python
   sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
   ```

4. **Box Plot**:
   - **Usage**: `sns.boxplot()`
   - **Use**: Box plots provide a graphical summary of a dataset based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. They are useful for identifying the distribution, central tendency, and variability of the data, as well as for detecting outliers.

   ```python
   sns.boxplot(x="day", y="total_bill", data=tips)
   ```

5. **Heatmap**:
   - **Usage**: `sns.heatmap()`
   - **Use**: Heatmaps display data in matrix form, with colors representing values. They are used to show the magnitude of a phenomenon as color in two dimensions, and are particularly useful for visualizing correlation matrices, frequency tables, and other forms of clustered data.

   ```python
   sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
   ```

Each of these plots serves different purposes and is useful for visualizing various types of data and relationships. Seaborn's high-level interface makes it easy to generate these plots with minimal code.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =

````python
import seaborn as sns
fmri=sns.load_dataset('fmri')
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Add title and labels
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
````

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram of 'price' with 'cut' as hue
sns.histplot(data=diamonds, x="price", hue="cut", bins=30, kde=True, palette='husl')

# Add title and labels
plt.title("Histogram of Price by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")

# Show the plot
plt.show()
````

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the flights dataset
flights = sns.load_dataset("flights")

# Create a pivot table
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Add title and labels
plt.title("Number of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
````
